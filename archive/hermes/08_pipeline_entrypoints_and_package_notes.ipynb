## Notebook/Script Bootstrap

Small helper formerly used to make imports work from command-line scripts.

_Archived from `scripts/_bootstrap.py`._


```python
"""Make the local src package importable when running scripts directly."""

import sys
from pathlib import Path


PROJECT_ROOT = Path(__file__).resolve().parents[1]
SRC_DIR = PROJECT_ROOT / "src"

if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))
```

## Full Pipeline Runner

Former entry point for running the complete scripted pipeline.

_Archived from `scripts/run_pipeline.py`._


```python
from _bootstrap import PROJECT_ROOT  # noqa: F401
from icd10_coding.ensemble.optimized import run as run_optimized
from icd10_coding.ensemble.meta_ensemble import run as run_meta_ensemble
from icd10_coding.models.chapter import run as run_chapter
from icd10_coding.models.classifier import run as run_classifier
from icd10_coding.processing.preprocess import run as run_preprocess
from icd10_coding.retrieval.matching import run as run_matching
from icd10_coding.retrieval.tfidf_retrieval import run as run_retrieval


def main() -> None:
    run_preprocess()
    run_matching()
    run_retrieval()
    run_classifier()
    run_optimized()
    run_chapter()
    run_meta_ensemble()


if __name__ == "__main__":
    main()
```
